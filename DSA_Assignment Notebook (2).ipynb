### 1.2 Visualisation

Use the artist layer of Matplotlib to replicate the bar chart below to visualize the percentage of the respondents' interest in the different data science topics surveyed.
To create this bar chart, you can follow the following steps:

1. Sort the dataframe in descending order of Very interested.
2. Convert the numbers into percentages of the total number of respondents. Recall that 2,233 respondents completed the survey. Round percentages to 2 decimal places.

As for the chart:
1. use a figure size of (20, 8),
2. bar width of 0.8,
3. use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars,

````python

#sort the dataframe based on the Very Interested values. The 'inplace=True' modifies the original dataframe.
#setting ascending to False, makes the values to be sorted in descending order.
df.rename(columns={'Unnamed: 0':'Topic'},inplace=True) #Rename the unnamed column to Topic
df_perc = df[['Topic']] #create a new dataframe containing the Topic column

#convert the values in the columns to percentage by using the formula
df_perc = df_perc.join((df[['Very interested','Somewhat interested','Not interested']]/2233)*100)

df_perc.round(2) #round the percentage vaues to 2 decimal places.
````

````python
#create the plot
````

````python
# Define a list of labels representing different areas of data science
labels = ['Data Analysis / Statistics', 'Machine Learning', 'Data Visualization', 'Big Data (Spark / Hadoop)', 'Deep Learning', 'Data Journalism']

# Extract the 'Very interested' column from the dataframe df_perc
very_int = df_perc['Very interested']
# Extract the 'Somewhat interested' column from the dataframe df_perc
some_int = df_perc['Somewhat interested']

# Extract the 'Not interested' column from the dataframe df_perc
not_int = df_perc['Not interested']

ind = np.arange(len(very_int))  # Create an array of indices for the x-axis based on the length of the 'very interested' data
width = 0.3  # Set the width of the bars in the bar chart

# Create a figure and a set of subplots with a specified size
fig, ax = plt.subplots(figsize=(20,8))

# Create a bar chart for the 'Very interested' data, offsetting the bars to the left
rects1 = ax.bar(ind - width, very_int, width, label='Very interested', color='#5cb85c')

# Create a bar chart for the 'Somewhat interested' data, placing the bars in the center
rects2 = ax.bar(ind, some_int, width, label='Somewhat interested', color='#5bc0de')

# Create a bar chart for the 'Not interested' data, offsetting the bars to the right
rects3 = ax.bar(ind + width, not_int, width, label='Notr interested', color='#d9534f')

ax.set_title("Percentage of Respondents' Interest In Data Science Areas", fontsize=16)  # Set the title of the plot with a specified font size
ax.set_xticks(ind)  # Set the x-ticks to the indices of the bars
ax.set_xticklabels((labels))  # Set the labels for the x-ticks using the provided labels
ax.get_yaxis().set_visible(False)  # Hide the y-axis for a cleaner look
ax.spines['top'].set_visible(False)  # Hide the top spine (border) of the plot
ax.spines['left'].set_visible(False)  # Hide the left spine (border) of the plot
ax.spines['right'].set_visible(False)  # Hide the right spine (border) of the plot
ax.legend(fontsize=14)  # Add a legend to the plot with a specified font size


def autolabel(rects, xpos='center'):
   
    # Define horizontal alignment options for the text labels
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    # Define offsets for the text labels based on their position
    offset = {'center': 0, 'right': 1, 'left': -1}

    # Iterate over each rectangle (bar) in the provided list
    for rect in rects:
        # Get the height of the bar and round it to 2 decimal places
        height = rect.get_height().round(2)
        # Annotate the bar with its height value
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),  # Position the text above the bar
                    xytext=(offset[xpos]*3, 3),  # Use 3 points offset for the text position
                    textcoords="offset points",  # Specify that the offset is in points
                    ha=ha[xpos], va='bottom', fontsize=14)  # Set horizontal and vertical alignment for the text

# Annotate the first set of bars (rects1) with their heights, positioning the labels at the center
autolabel(rects1, "center")

# Annotate the second set of bars (rects2) with their heights, positioning the labels at the center
autolabel(rects2, "center")

# Annotate the third set of bars (rects3) with their heights, positioning the labels at the center
autolabel(rects3, "center")

# Adjust the layout of the figure to prevent overlap and ensure a clean presentation
fig.tight_layout()

# Display the plot to the user
plt.show()
````

````python
# One-hot encode the "Category" variable and add the encoded feature to the dataframe 
# Name the new columns (hint: columns = ohe.get_feature_names())

# Initialize the OneHotEncoder
ohe = OneHotEncoder(sparse=False, drop='first')  # Use drop='first' to avoid the dummy variable trap

# Fit and transform the 'Category' column to one-hot encode it
encoded_categories = ohe.fit_transform(df[['Category']])  # Fit the encoder to the 'Category' column and transform it

# Create a DataFrame from the encoded categories
encoded_df = pd.DataFrame(encoded_categories, columns=ohe.get_feature_names_out(['Category']))  # Get the new column names



# Add encoded feature to the dataframe and view the new dataframe with .head()

df = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)  # Reset index to avoid misalignment

print(df.head())  # Display the first few rows of the updated DataFrame
````

#### 2.5 Split the data into a training and test set

Recall that, in the 3.3 IDE activity M3U2 practice notebook, the steps in the following cell were introduced. You may also reference the documentation on the [scikit-learn website](https://scikit-learn.org/stable/index.html) for the following:
* train_test_split()
* DecisionTreeRegressor()
* mean_squared_error()

````python
#from typing import Tuple

def train_decision_tree_regressor(X: pd.DataFrame, y: pd.Series) -> Tuple[DecisionTreeRegressor, float]:
   
    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

    # Print the shapes of the training and test sets
    print("Training Features (X_train) shape:", X_train.shape)
    print("Test Features (X_test) shape:", X_test.shape)
    print("Training Response (y_train) shape:", y_train.shape)
    print("Test Response (y_test) shape:", y_test.shape)

    # Initialize the Decision Tree Regressor
    model = DecisionTreeRegressor(random_state=0)

    # Fit the model to the training data
    model.fit(X_train, y_train)

    # Print the model's parameters to verify it has been fitted
    print("Model fitted with parameters:", model.get_params())

    # Make predictions on the training set
    y_pred_train = model.predict(X_train)

    # Calculate the Mean Squared Error (MSE)
    error = mean_squared_error(y_train, y_pred_train)

    # Print the MSE
    print("Mean Squared Error (MSE) prior to pruning:", error)

    return model, error
````
